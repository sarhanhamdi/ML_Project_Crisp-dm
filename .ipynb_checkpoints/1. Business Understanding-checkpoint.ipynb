# Phase 1 : Business Understanding

The Uber dataset contains details about every trip, such as where the ride started and ended, the time of the trip, the number of passengers, and the cost. The purpose of using this data is to help Uber in two ways: first, by predicting how much a ride will cost, and second, by understanding patterns in trips to improve services and decision-making.

## 1. Business Objective 1 :
Uber desires to provide accurate cost estimates to riders before booking a trip. This improves customer trust and satisfaction as there is no unexpected costs, and helps riders make informed decisions.
## 1.1 Data Science Objective 1 (Supervised) :
Developping a supervised model to predict the variable of the ride cost based on the features of the Dataset.

## 2. Business Objective 2 :
Uber also wants to gain knowledge about ride behavior by identifying patterns and grouping similar trips. This helps to understand the market segmentation and detect anomalies or fraud.

## 2.1 Data Science Objective 2 (Unsupervised) :
Apply unsupervised learning techniques (clustering) to group trips based on features such as trip distance, time of day, and passenger count. The goal is to reveal hidden structures in the data, detect outliers, and provide actionable business insights that go beyond fare prediction.
